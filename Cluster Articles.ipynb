```python
articles = requests.get('http://www.usatoday.com/news_sitemap_news_politics.xml')
```

```python
from xml.etree import ElementTree
tree = ElementTree.fromstring(articles.content)
```

```python
articles = []
for node in tree:
    article = Article(node[0].text)
    article.download()
    article.parse()
    articles.append(article.text)
```

```python
len(articles)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

num_clusters = 8
vectorizer = TfidfVectorizer(max_df=0.5, max_features=None,
                                 min_df=2, stop_words='english',
                                 use_idf=False)
X = vectorizer.fit_transform(articles)

from sklearn.cluster import MiniBatchKMeans
km = MiniBatchKMeans(n_clusters=num_clusters, init='k-means++', n_init=1,
                         init_size=1000, batch_size=1000, verbose=False)
km.fit(X)
```

```python
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
```

```python
order_centroids
```

```python
terms = vectorizer.get_feature_names()
```

```python
terms
```

```python
for i in range(8):
        print "Cluster %d:" % i
        for ind in order_centroids[i, :10]:
            print ' %s' % terms[ind]
        print()
```
