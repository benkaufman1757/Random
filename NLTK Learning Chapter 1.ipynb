```python
# for python2.x
#from __future__ import division

# How diverse a text is, (number of distinct words used / number of words used)
def lexical_diversity(text):
    return len(set(text)) / len(text)

# Helper function
def percentage(count, total):
    return 100 * count / total
```

```python
lexical_diversity(text3)
```

```python
# Frequency distribtion of words (how many times a word appears in a text)
fdist1 = FreqDist(text1)
```

```python
fdist1.most_common(20)
```

```python
# Plot the frequency distribution
fdist1.plot(50)
```

```python
# Cumulative plot
fdist1.plot(50, cumulative=True)
```

```python
# Create a frequency distribution of lengths of words in a text
fdist = FreqDist(len(w) for w in text1)
```

```python
fdist.most_common()
```

```python
# Most frequent word length
fdist.max()
```

```python
# Percent of words with a word length of 3
fdist.freq(3)
```
