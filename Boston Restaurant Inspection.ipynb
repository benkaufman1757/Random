```python
employee_earnings = pd.read_csv('/Users/benkaufman/Downloads/Employee_Earnings_Report_2014.csv')
```

```python
employee_earnings.head()
```

```python
employee_earnings.loc[employee_earnings['DEPARTMENT NAME'].str.contains('Office')]['DEPARTMENT NAME'].value_counts()
```

```python
employee_earnings.dtypes
```

```python
employee_earnings['TOTAL EARNINGS'] = employee_earnings['TOTAL EARNINGS'].apply(lambda x: x[1:])
```

```python
employee_earnings['TOTAL EARNINGS'] = employee_earnings['TOTAL EARNINGS'].astype(float)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.figure(figsize=(30,10))
grouped = employee_earnings.groupby('DEPARTMENT NAME')['TOTAL EARNINGS'].aggregate(sum)
sns.barplot(x=grouped.index, y=grouped)
locs, labels = sns.plt.xticks()
sns.plt.setp(labels, rotation=90)
sns.plt.show()
```
