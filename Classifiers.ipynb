```python
import os
politics_dir = '/Users/benkaufman/Downloads/SomasundaranWiebe-politicalDebates/'
politics_file_ids = []
for x in os.listdir(politics_dir):
    if os.path.isdir(politics_dir+x):
        for y in os.listdir(politics_dir+x):
            try:
                f = open(politics_dir+x+'/'+y, 'r')
                lines = f.readlines()
                f.close()
                f = open(politics_dir+x+'/'+y, 'w')
                f.writelines(lines[3:])
                f.close()
                politics_file_ids.append(x+'/'+y)
                print(x+'/'+y)
            except:
                pass
                #print(x+'/'+y)
```

```python
politics_file_ids
```

```python
from nltk.corpus import PlaintextCorpusReader
politics = PlaintextCorpusReader(politics_dir, politics_file_ids)
```

```python
dir(politics)
```

```python
os.listdir(politics_dir+'abortion')
```

```python
politics_target = ['abortion', 'gayRights', 'creation', 'god', 'healthcare', 'guns']
politics_data = []
for x in politics.fileids():
    if 'abortion' in x:
        politics_data.append((politics.raw(x), 'abortion'))
    elif 'gayRights' in x:
        politics_data.append((politics.raw(x), 'gayRights'))
    elif 'creation' in x:
        politics_data.append((politics.raw(x), 'creation'))
    elif 'god' in x:
        politics_data.append((politics.raw(x), 'god'))
    elif 'healthcare' in x:
        politics_data.append((politics.raw(x), 'healthcare'))
    elif 'guns' in x:
        politics_data.append((politics.raw(x), 'guns'))
```

```python
from random import shuffle
shuffle(politics_data)
politics_train_data, politics_targets = zip(*politics_data[:4500])
politics_score_data, politics_score_targets = zip(*politics_data[4500:])
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()

X_politics_train_counts = count_vect.fit_transform(politics_train_data)
print(X_politics_train_counts.shape)

from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_politics_train_counts)
X_politics_train_tf = tf_transformer.transform(X_politics_train_counts)
print(X_politics_train_tf.shape)

tfidf_transformer = TfidfTransformer()
X_politics_train_tfidf = tfidf_transformer.fit_transform(X_politics_train_counts)

from sklearn.naive_bayes import MultinomialNB
politics_clf = MultinomialNB(alpha=0.01).fit(X_politics_train_tfidf, politics_targets)
```

```python
politics_stuff = ['Senate Majority Leader Bill Frist said Friday that "it is impossible" to have all Americans covered by health insurance, but he predicted that Congress would take incremental steps to expand coverage this year. The Tennessee Republican said his state was "going bankrupt" as a result of trying to achieve universal insurance coverage. "It is impossible to get everybody covered," he said at a meeting with journalists. "Its impossible to get to 100 percent."', 'on the third day god created the rainforest']
X_politics_new_counts = count_vect.transform(politics_stuff)

X_politics_new_tfidf = tfidf_transformer.transform(X_politics_new_counts)

politics_predicted = politics_clf.predict_log_proba(X_politics_new_tfidf)
print(politics_clf.predict(X_politics_new_tfidf))
print(politics_predicted)
```
