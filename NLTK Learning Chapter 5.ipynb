```python
# Most frequent nouns by part of speech
def findtags(tag_prefix, tagged_text):
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text
                                  if tag.startswith(tag_prefix))
    return dict((tag, cfd[tag].most_common(5)) for tag in cfd.conditions())

tagdict = findtags('NN', nltk.corpus.brown.tagged_words(categories='news'))
for tag in sorted(tagdict):
    print(tag, tagdict[tag])
    
# -HL = headlines
# -TL = titles
# -NC = Citations
#   $ = possesive noun
#   S = plural noun
```

```python
# Create frequency distribution of words that follow the word "often"
brown_lrnd_tagged = brown.tagged_words(categories='learned', tagset='universal')
tags = [b[1] for (a, b) in nltk.bigrams(brown_lrnd_tagged) if a[0] == 'often']
fd = nltk.FreqDist(tags)
fd.tabulate()
```

```python
# find trigrams that fit the model (Verb, "to", Verb)
from nltk.corpus import brown
def process(sentence):
    for (w1,t1), (w2,t2), (w3,t3) in nltk.trigrams(sentence):
        if (t1.startswith('V') and t2 == 'TO' and t3.startswith('V')):
            print(w1, w2, w3)
for tagged_sent in brown.tagged_sents():
    process(tagged_sent)
```

```python
# Get most ambiguous words with regards to POS
brown_news_tagged = brown.tagged_words(categories='news', tagset='universal')
data = nltk.ConditionalFreqDist((word.lower(), tag) for (word, tag) in brown_news_tagged)

for word in sorted(data.conditions()):
    if len(data[word]) > 3:
        tags = [tag for (tag, _) in data[word].most_common()]
        print(word, ' '.join(tags))
```

```python
from nltk.corpus import brown
brown_tagged_sents = brown.tagged_sents(categories='news')
brown_sents = brown.sents(categories='news')
```

```python
# Simple tagger assigns the same tag to each token
tags = [tag for (word, tag) in brown.tagged_words(categories='news')]
nltk.FreqDist(tags).max()
```

```python
# Tag everything as a noun
raw = 'I do not like green eggs and ham, I do not like them Sam I am!'
tokens = word_tokenize(raw)
default_tagger = nltk.DefaultTagger('NN')
default_tagger.tag(tokens)
```

```python
# Evaluate performance of tagger 1=100% correct
default_tagger.evaluate(brown_tagged_sents)
```

```python
# Regex tagger
patterns = [
     (r'.*ing$', 'VBG'),               # gerunds
     (r'.*ed$', 'VBD'),                # simple past
     (r'.*es$', 'VBZ'),                # 3rd singular present
     (r'.*ould$', 'MD'),               # modals
     (r'.*\'s$', 'NN$'),               # possessive nouns
     (r'.*s$', 'NNS'),                 # plural nouns
     (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),  # cardinal numbers
     (r'.*', 'NN')                     # nouns (default)
]
regexp_tagger = nltk.RegexpTagger(patterns)
regexp_tagger.tag(brown_sents[3])
```

```python
# Evaluate regex tagger
regexp_tagger.evaluate(brown_tagged_sents)
```

```python
# Lookup Tagger
fd = nltk.FreqDist(brown.words(categories='news'))
cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))

# Get top 100 most frequent words and use those tags
most_freq_words = fd.most_common(100)

# Create dict of most common words to pass into tagger
likely_tags = dict((word, cfd[word].max()) for word, _ in most_freq_words)
baseline_tagger = nltk.UnigramTagger(model=likely_tags)

# evaluate lookup tagger
baseline_tagger.evaluate(brown_tagged_sents)
```

```python
# Try to tag sentences from brown news with the lookup tagger
sent = brown.sents(categories='news')[3]
baseline_tagger.tag(sent)
```

```python
# Specify backoff tagger in case the word is not found in the 100 most common (lookup tagger)
baseline_tagger = nltk.UnigramTagger(model=likely_tags,
                                     backoff=nltk.DefaultTagger('NN'))
```

```python
%matplotlib inline
# Evaluate performance by with varying sizes of wordlists (lookup lists)
def performance(cfd, wordlist):
    lt = dict((word, cfd[word].max()) for word in wordlist)
    baseline_tagger = nltk.UnigramTagger(model=lt, backoff=nltk.DefaultTagger('NN'))
    return baseline_tagger.evaluate(brown.tagged_sents(categories='news'))

def display():
    import pylab
    words_by_freq = list(nltk.FreqDist(brown.words(categories='news')))
    cfd = nltk.ConditionalFreqDist(brown.tagged_words(categories='news'))
    sizes = 2 ** pylab.arange(15)
    perfs = [performance(cfd, words_by_freq[:size]) for size in sizes]
    pylab.plot(sizes, perfs, '-bo')
    pylab.title('Lookup Tagger Performance with Varying Model Size')
    pylab.xlabel('Model Size')
    pylab.ylabel('Performance')
    pylab.show()
display()
```

```python
# Training and testing taggers 90%-10% train-test data 
size = int(len(brown_tagged_sents) * 0.9)
train_sents = brown_tagged_sents[:size]
test_sents = brown_tagged_sents[size:]
unigram_tagger = nltk.UnigramTagger(train_sents)
unigram_tagger.evaluate(test_sents)
```

```python
# n-gram taggers evaluate the current word with the POS tag and does the same with the n-1 preceeding terms
# chooses most likely POS tag 
bigram_tagger = nltk.BigramTagger(train_sents)
bigram_tagger.tag(brown_sents[2007])
```

```python
unseen_sent = brown_sents[4203]
bigram_tagger.tag(unseen_sent)
```

```python
bigram_tagger.evaluate(test_sents)
```

```python
# the sparse data problem is tradeoff between precision and recall
# the bigram tagger has very bad performance on the unseen sentence because a lot of the words it sees are unknown
# Also note n-gram taggers should not go across multiple sentences
```

```python
# You can combine taggers with the backoff parameter
t0 = nltk.DefaultTagger('NN')
t1 = nltk.UnigramTagger(train_sents, backoff=t0)
t2 = nltk.BigramTagger(train_sents, backoff=t1) # Set the cuttoff to 2 if you want to get rid of words 
                                                # it has only seen 1 or 2 times
t2.evaluate(test_sents)
```

```python
# When it comes to tagging unkwown words, default taggers will make it a noun most likely but n-gram taggers
# will see if the word preceeding the unknown word is "to" and will probably define it as a verb
```

## Transformation-Based Tagging

* n-gram tagging will not work across languages
* requires substantial data for the tagger's model

Transformational-Based tagging attempts to construct rules for languages based on patterns from both POS tags
as well as the word itself.

Factors:
* Morphological Clues: adj + -ness = noun ex) happy -> happiness
* Syntactic Clues: adjective preceeded by "very" ex) The window is very near. near = adjective
* Semantic Clues: across languages if words are in the same position it is possible that they are of the same POS
* New Words: nouns are an "open class" whereas prepositions are a "closed class" because you can't create a new one
* Morphology in Part of Speech Tagsets: verbs can have different forms for different tenses that are constant
